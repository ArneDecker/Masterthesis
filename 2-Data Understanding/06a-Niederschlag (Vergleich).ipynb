# <font size="7">Vergleich des Niederschlags</font>

<font size="5">Wie auch bei den anderen Merkmalen ist der Verlauf des Niederschlags in den Städten sehr ähnlich. Die Ausschläge nach oben geschehen zeitlich parallel, sind allerdings stellenweise etwas unterschiedlich stark. Meist liegt Ulm deutlich über dem Durchschnitt, Mannheim deutlich darunter und die anderen Städte bewegen sich in etwa auf dem Durchschnitt.</font>

```python
#Daten aus jeweiliger CSV-Datei laden
df = pd.read_csv("Datenbeschaffung/kalender.csv", index_col=0, parse_dates=True, usecols=[0,2])
df = df.join(pd.read_csv("Datenbeschaffung/stuttgart.csv", index_col=0, parse_dates=True, usecols=[0,10], squeeze=True).rename("stuttgart"))
df = df.join(pd.read_csv("Datenbeschaffung/freiburg.csv", index_col=0, parse_dates=True, usecols=[0,10], squeeze=True).rename("freiburg"))
df = df.join(pd.read_csv("Datenbeschaffung/mannheim.csv", index_col=0, parse_dates=True, usecols=[0,10], squeeze=True).rename("mannheim"))
df = df.join(pd.read_csv("Datenbeschaffung/ulm.csv", index_col=0, parse_dates=True, usecols=[0,10], squeeze=True).rename("ulm"))

#Durchschnitt berechnen
df["durchschnitt"] = round(((df["stuttgart"] + df["freiburg"] + df["mannheim"] + df["ulm"]) / 4), 1)

print(df)
```

```python
#Jahre einzeichnen
functions.draw_years(df)

#Daten für jede Stadt und Durchschnitt plotten
plt.plot(df["durchschnitt"], label="Durchschnitt")
plt.plot(df["stuttgart"], label="Stuttgart");
plt.plot(df["freiburg"], label="Freiburg");
plt.plot(df["mannheim"], label="Mannheim");
plt.plot(df["ulm"], label="Ulm");

#Achsen und Layout
plt.title("Niederschlag in allen Städten (2015 - 2021)")
plt.ylabel("Niederschlag in Millimetern/Quadratmeter")
plt.ylim(-5, 100)
plt.xlabel("Datum")
plt.xlim(dt.datetime(2015,1,1), dt.datetime(2021,12,31))

#Legende einfügen
handles, labels = plt.gca().get_legend_handles_labels()
by_label = OrderedDict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys(), loc="lower left", bbox_to_anchor=(0,-0.15), ncol=4)

plt.show()

#Deskriptive Statistiken
print("\n")
print(df.iloc[:,1:6].describe().transpose())
print("\n")
```

```python
for year in [2015, 2016, 2017, 2018, 2019, 2020, 2021]:
    
    print("\n\n", str(year), "--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------\n\n")
    
    #Untere und obere Grenze zum Filtern des DataFrames nach jeweiligem Jahr
    lower_border = str(year) + "-01-01"
    upper_border = str(year) + "-12-31"
    
    #Wochen einzeichnen
    functions.draw_weeks(df[lower_border:upper_border])
    
    #Daten für jede Stadt und Durchschnitt plotten
    plt.plot(df[lower_border:upper_border]["durchschnitt"], label="Durchschnitt")
    plt.plot(df[lower_border:upper_border]["stuttgart"], label="Stuttgart");
    plt.plot(df[lower_border:upper_border]["freiburg"], label="Freiburg");
    plt.plot(df[lower_border:upper_border]["mannheim"], label="Mannheim");
    plt.plot(df[lower_border:upper_border]["ulm"], label="Ulm");
    
    #Achsen und Layout
    plt.title("Niederschlag in allen Städten (" + str(year) + ")")
    plt.ylabel("Niederschlag in Millimetern/Quadratmeter")
    plt.ylim(-5, 100)
    plt.xlabel("Datum")
    plt.xlim(dt.datetime(year,1,1), dt.datetime(year,12,31))
    
    #Legende einfügen
    handles, labels = plt.gca().get_legend_handles_labels()
    by_label = OrderedDict(zip(labels, handles))
    plt.legend(by_label.values(), by_label.keys(), loc="lower left", bbox_to_anchor=(0,-0.15), ncol=4)

    plt.show()
    
    #Deskriptive Statistiken
    print("\n")
    print(df[lower_border:upper_border].iloc[:,1:6].describe().transpose())
    print("\n")
```

<font size="5">Bei dieser Analyse wird außerdem explizit nach Tagen gesucht, an denen es in einigen Städten regnet und in anderen nicht. Dafür wird je Stadt ein Indikator eingefügt, der ab einem Niederschlag von 0,5 Millimetern pro Quadratmeter auf 1 gesetzt wird, an übrigen Tagen mit fast oder gar keinem Niederschlag wird der Indikator auf 0 gesetzt. Anschließend wird nach Tagen gefiltert, bei denen sich die Indikatoren der Städte unterscheiden. An solchen Tagen regnet es in machen Städten und in andern nicht. Da sich keine solchen Tage in den Daten finden lassen, kann man davon ausgehen, dass der Niederschlag in allen Städten zwar unterschiedlich stark ausgeprägt ist, allerdings trotzdem stets synchron einsetzt.</font>

```python
#Indikator für Regen einfügen
df["stuttgart_regen"] = np.where(df["stuttgart"] < 0.5, 1, 0)
df["freiburg_regen"] = np.where(df["stuttgart"] < 0.5, 1, 0)
df["mannheim_regen"] = np.where(df["stuttgart"] < 0.5, 1, 0)
df["ulm_regen"] = np.where(df["stuttgart"] < 0.5, 1, 0)

print(df)

print(df[(df["stuttgart_regen"] != df["freiburg_regen"]) & (df["freiburg_regen"] != df["mannheim_regen"]) & (df["mannheim_regen"] != df["ulm_regen"])])
```

<font size="5">Die Korrelationskoeffizienten zeigen, dass die vier Städte relativ stark mit dem Durchschnitt korrelieren. Die Korrelation ist nicht so stark ausgeprägt wie bei der Temperatur, allerdings ist sie trotzdem signifikant und ausreichend stark. Lediglich Mannheim korreliert nur mittelstark mit dem Durchschnitt. Das Regenwetter ist trotzdem in allen Städten synchron. Daher kann der Durchschnitt über alle Städte für die weiteren Analysen verwendet werden.</font>

```python
#Heatmap der Korrelationen untereinander (Pearson)
ax = sns.heatmap(data=df.corr(method="pearson"), annot=True, vmin=0, vmax=1, cmap="rocket_r");
ax.set_title("Pearson")
ax.xaxis.set_ticks_position("top")

plt.show()

#Heatmap der Korrelationen untereinander (Spearman)
ax = sns.heatmap(data=df.corr(method="spearman"), annot=True, vmin=0, vmax=1, cmap="rocket_r");
ax.set_title("Spearman")
ax.xaxis.set_ticks_position("top")

plt.show()

#Heatmap der Korrelationen untereinander (Kendall)
ax = sns.heatmap(data=df.corr(method="kendall"), annot=True, vmin=0, vmax=1, cmap="rocket_r");
ax.set_title("Kendall")
ax.xaxis.set_ticks_position("top")

plt.show()
```
